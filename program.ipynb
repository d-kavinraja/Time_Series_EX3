```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = [3, 16, 156, 47, 246, 176, 233, 140, 130, 101, 166, 201, 200, 116, 118, 247, 209, 52, 153, 232, 128, 27, 192, 168, 208, 187, 228, 86, 30, 151, 18, 254, 76, 112, 67, 244, 179, 150, 89, 49, 83, 147, 90, 33, 6, 158, 80, 35, 186, 127]
```

```python

# Mean
data_mean = np.mean(data)

# Variance
data_var = np.var(data)

# Normalized data
normalized_data = (data - data_mean) / np.sqrt(data_var)

# Compute the autocorrelation function (ACF)
acf_result = np.correlate(normalized_data, normalized_data, mode='full')

# Take only the positive lags
acf_result = acf_result[len(acf_result)//2:]
```

```python
# Plotting the Autocorrelation Function (ACF)
plt.figure(figsize=(10, 5))  # Setting the figure size
plt.stem(acf_result[:36], use_line_collection=True, linefmt='blue', markerfmt='b.', basefmt='b')  # Stem plot with blue color
plt.xlabel('Lag')  # Label for x-axis
plt.ylabel('Autocorrelation')  # Label for y-axis
plt.title('Autocorrelation Function (ACF)')  # Title of the plot
plt.show()  # Displaying the plot
```
